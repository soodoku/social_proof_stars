### Treat the Treatment Packages

```python
import subprocess
import pandas as pd

# List of treated packages to download
df = pd.read_csv("data/fin_dataset_treat_control.csv")
df_treat = df[df['treatment'] == 1]
df_treat
```

```python
treat_packages = df_treat['file_project'].to_list()

# Number of downloads per package
num_downloads = 100

# Loop through each package and download it multiple times
for package in treat_packages:
    for _ in range(num_downloads):
        # Install the package
        subprocess.call(['pip', 'install', '--no-cache-dir', package])
        
        # Uninstall the package
        subprocess.call(['pip', 'uninstall', '-y', package])
```
