```python
import json
from datetime import datetime

import pandas as pd
import requests
import yaml
from tqdm.notebook import tqdm
from utilities import read_jsons

today = datetime.today()
today_str = today.strftime('%Y-%m-%d')
today_str
```

```python
with open("../input/stargazers.yaml", 'r') as file:
    users = yaml.safe_load(file).split()
users
```

```python
# Get profile using GitHub API
payload_savepath = "../output/stargazers_profile_payload"
if False:
    payload_data = []
    for user in tqdm(users):
        # Query
        url_endpoint = f"https://api.github.com/users/{user}"
        response = requests.get(url_endpoint)
        rc = response.status_code
        assert rc == 200
        payload = response.json()

        # Store to disk
        with open(f"{payload_savepath}/{user}_{today_str}.json", 'w') as file:
            json.dump(payload, file)

        # Append
        payload_data.append(payload)
        
payload_data = read_jsons(payload_savepath)
payload_data[:1]
```

```python
df = (
    pd.DataFrame(payload_data)
    .assign(retrieval_date=today)
)
df.to_csv("../output/stargazers_profile.csv", index=False)
df
```
