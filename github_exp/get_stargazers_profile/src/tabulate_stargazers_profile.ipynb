```python
import pandas as pd

pd.set_option("display.max_columns", 20)
from tableone import TableOne
from utilities import pandas_to_tex
import warnings

warnings.filterwarnings("ignore")

df = (
    pd.read_csv("../output/stargazers_profile.csv")
    .drop_duplicates("id")
    .reset_index(drop=True)
    .assign(
        created_at=lambda df_: pd.to_datetime(df_["created_at"]).dt.tz_localize(None),
        retrieval_date=lambda df_: pd.to_datetime(df_["retrieval_date"]).dt.tz_localize(
            None
        ),
        age=lambda df_: df_["retrieval_date"] - df_["created_at"],
        age_yrs=lambda df_: [timedelta.days / 365.25 for timedelta in df_["age"]],
    )
    .assign(
        **{
            k: lambda df_, col=k: df_[col[1:]].notna()
            for k in [
                "Iname",
                "Icompany",
                "Iblog",
                "Ilocation",
                "Ibio",
                "Itwitter_username",
            ]
        }
    )
)
df
```

```python
columns = [
    "public_repos",
    "public_gists",
    "followers",
    "following",
    "age_yrs",
    "Iname",
    "Icompany",
    "Iblog",
    "Ilocation",
    "Ibio",
    "Itwitter_username",
]
labels = dict(
    public_repos="Public repositories",
    public_gists="Public gists",
    followers="Followers",
    following="Following",
    age_yrs="Account age (years)",
    Iname="Lists name",
    Icompany="Lists company",
    Iblog="Lists blog/website",
    Ilocation="Lists geographical location",
    Ibio="Lists brief biography",
    Itwitter_username="Lists Twitter handle",
)
```

```python
means = TableOne(
    data=df,
    columns=columns,
    rename=labels,
    label_suffix=False,
    missing=False,
)
means
```

```python
medians = TableOne(
    data=df,
    columns=columns,
    rename=labels,
    label_suffix=False,
    missing=False,
    nonnormal=columns,
)
medians
```

```python
df_tabulate_stats = pd.concat(
    [means.tableone, medians.tableone], ignore_index=True, axis=1
).iloc[1:]
pandas_to_tex(
    df_tabulate_stats, "../output/organic_starrers_profile_summary", index=True
)
df_tabulate_stats
```
