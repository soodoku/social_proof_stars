```python
import requests

package = "pandas"
```

```python
# Make a GET request to the PyPI API endpoint for the package
response = requests.get(f"https://pypi.org/pypi/{package}/json")

if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    package_info = {
        "name": data["info"]["name"],
        "version": data["info"]["version"],
        "release_date": data["urls"][0]["upload_time"],
        "homepage": data["info"]["home_page"],
        "description": data["info"]["summary"],
        "author": data["info"]["author"],
        "author_email": data["info"]["author_email"],
        "license": data["info"]["license"],
    }
    print("Package information:")
    print("Name:", package_info["name"])
    print("Version:", package_info["version"])
    print("Release Date:", package_info["release_date"])
    print("Homepage:", package_info["homepage"])
    print("Description:", package_info["description"])
    print("Author:", package_info["author"])
    print("Author Email:", package_info["author_email"])
    print("License:", package_info["license"])
else:
    print("Failed to retrieve package information. Status code:", response.status_code)
```
