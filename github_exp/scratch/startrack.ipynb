```python
import requests
import json

# OWNER = "lsys"
# REPO = "forestplot"
OWNER = "lsys"
REPO = "forestplot"

url = f"https://api.github.com/repos/{OWNER}/{REPO}/stargazers"

headers = {"Accept": "application/vnd.github.v3.star+json"}
params = {"per_page": 100, "page": 1}
```

```python
response = requests.get(url, headers=headers, params=params)

response.status_code
```

```python
len(response.json())
```

```python
response.json()
```

```python
with open(f"startrack.json", 'w') as fh:
    json.dump(response.json(), fh) 
```

```python
with open(f"startrack.json", 'r') as fh:
    stars = json.load(fh) 
stars
```
